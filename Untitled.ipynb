```python
import numpy as np
import matplotlib.pyplot as plt
import time

def pltsin(ax):
    x = np.linspace(0,1,100)
    y = np.random.random(size=(100,1))
    if len(ax.lines):
        ax.lines[0].set_xdata(x)
        ax.lines[0].set_ydata(y)
    else:
        ax.plot(x,y)
    fig.canvas.draw()

fig,ax = plt.subplots(1,1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_xlim(0,1)
ax.set_ylim(0,1)
for f in range(60):
    pltsin(ax)
    time.sleep(.1)
```

```python
# metropolis algorithm for discrete distribution and
# explicit logic

import seaborn
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randn, randint, uniform, choice
def metropolis_d1(a, trials, burn_in):
    N = len(a)
    counts = np.zeros(N)
    idx = randint(0, N)  # current position

    for i in range(trials + burn_in):
        move = (idx + choice([-1, 1])) % N
    
        if a[move] >= a[idx]: #always move if new population is higher
            idx = move
        else:
            # probabilisitically move based on ratio of new/current population
            if uniform() < a[move] / a[idx]:
                idx = move

        if i >= burn_in:
            counts[idx] += 1
            step = i - burn_in + 1
            if (step < 100 and step% 2 ==0) or (step < 200 and step % 10 ==0) or step % 200 == 0:
                plt.bar(range(len(counts)), counts);
                plt.ylim(0, 2000)
                fig.canvas.draw()
                plt.title('step=' + str(step))
    return counts/sum(counts)


fig = plt.figure()
a = np.array([10, 10, 10, 30, 60, 90, 120, 20])
counts = metropolis_d1(a, 5000, 500)
plt.bar(range(len(counts)), counts);
```
